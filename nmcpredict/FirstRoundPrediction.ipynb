<h3>SPLIT TABLES<H3>

```python
mini_tables = {}
counter = 0
for i in subjects:
    j = counter
    try:
        mini_tables[i] = abia_clean[header_indices[j]+1:header_indices[j+1]]
    except:
        mini_tables[i] = abia_clean[header_indices[j]+1:]
    counter +=1
```

```python
mini_tables[subjects[5]]
```

<h3>ADD SUBJECT ROW<h3>

```python
for table_name, table in mini_tables.items():
    table['SUBJECT'] = table_name
```

```python
mini_tables['SENIOR MATHEMATICS']
```

```python
tables = [t for t in mini_tables.values()]
```

```python
final = pd.concat(tables, axis=0)
```

```python
final.describe()
```

```python
dummies=pd.get_dummies(final[['REMARKS', 'SUBJECT']], drop_first=True)
```

```python
final[dummies.columns] = dummies
```

```python
final.drop(['REMARKS', 'SUBJECT'], axis=1, inplace=True)
```
